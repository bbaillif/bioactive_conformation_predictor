```python
import os
import json
```

```python
gold_dir = 'gold_docking_pdbbind_random_0/'
for pdb_id in os.listdir(gold_dir) :
    result_path = os.path.join(gold_dir, pdb_id, 'results.json')
    if os.path.exists(result_path) :
        with open(result_path, 'r') as f :
            results = json.load(f)
            current_pdb_id = pdb_id
            if 'top_ligand_rmsd' in results['flexible'] :
                flexible_rmsd = results['flexible']['top_ligand_rmsd']
                rigid_rmsd = results['rigid']['ligand_rmsd_top_score']
                if (rigid_rmsd < flexible_rmsd) and (rigid_rmsd < 2) and (flexible_rmsd > 2) :
                    print(pdb_id)
                    print(f'Rigid: {rigid_rmsd}')
                    print(f'Flexible: {flexible_rmsd}')
```
