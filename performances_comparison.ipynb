```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s_data = pd.read_csv('../Downloads/scaffold_split.csv')
s_h_data = pd.read_csv('../Downloads/scaffold_split_h.csv')
r_data = pd.read_csv('../Downloads/random_split.csv')
r_h_data = pd.read_csv('../Downloads/random_split_h.csv')
```

```python
plt.figure(figsize=(10, 8))
plt.plot(s_data['Step'], s_data['Value'], label='Scaffold split')
plt.plot(s_h_data['Step'], s_h_data['Value'], label='Scaffold split with hydrogens')
plt.plot(r_data['Step'], r_data['Value'], label='Random split')
plt.plot(r_h_data['Step'], r_h_data['Value'], label='Random split with hydrogens')
plt.legend()
#plt.show()
plt.savefig('Performance_comparison_scaffold_random.png', dpi=300)
```
